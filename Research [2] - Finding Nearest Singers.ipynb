```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.preprocessing import Normalizer, StandardScaler, MaxAbsScaler, MinMaxScaler, RobustScaler, PowerTransformer, QuantileTransformer
```

```python
singers = pd.read_csv('singers_dataset.csv', index_col=0)
singers.iloc[:, 3:] = StandardScaler().fit_transform(singers.iloc[:, 3:])
singers.head()
```

```python
def find_nearest_singers(singer, metric, k=10):
    dist_func = euclidean_distances if metric == 'euclidean' else cosine_similarity
    ascending = True  if metric == 'euclidean' else False

    singers_similarity = pd.DataFrame(dist_func(singers.iloc[:, 3:]))
    singer_index = singers[singers['Singer'] == singer].index[0]
    top_similar = singers_similarity[singer_index].sort_values(ascending=ascending)[1:k+1]
    top_similar_names = list(map(lambda x: singers.loc[x]['Singer'], top_similar.index.values))
    top_similar_values = map(lambda x: round(x*100, 2), top_similar.values)

    print('\n'.join([f'{name} - {values}' for name, values in zip(top_similar_names, top_similar_values)]))
```

```python
singer = 'Eminem'
```

```python
find_nearest_singers(singer, 'cosine')
```

```python
find_nearest_singers(singer, 'euclidean')
```
