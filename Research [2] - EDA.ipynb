```python
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

from sklearn.preprocessing import Normalizer, StandardScaler, MaxAbsScaler, MinMaxScaler, RobustScaler

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
authors = pd.read_csv('authors_dataset.csv', index_col=0)
authors.iloc[:, 3:] = StandardScaler().fit_transform(authors.iloc[:, 3:])
authors.head()
```

```python
def find_nearest_authors_cosine(author, k=10):
    authors_similarity = pd.DataFrame(cosine_similarity(authors.iloc[:, 3:]))
    author_index = authors[authors['Author'] == author].index[0]
    top_similar = authors_similarity[author_index].sort_values(ascending=False)[1:k+1]
    top_similar_names = list(map(lambda x: authors.loc[x]['Author'], top_similar.index.values))
    top_similar_values = map(lambda x: round(x, 4) * 100, top_similar.values)
    name_value_pairs = list(zip(top_similar_names, top_similar_values))
    return name_value_pairs

def find_nearest_authors_euclidean(author, k=10):
    authors_similarity = pd.DataFrame(euclidean_distances(authors.iloc[:, 3:]))
    author_index = authors[authors['Author'] == author].index[0]
    top_similar = authors_similarity[author_index].sort_values(ascending=True)[1:k+1]
    top_similar_names = list(map(lambda x: authors.loc[x]['Author'], top_similar.index.values))
    top_similar_values = map(lambda x: round(x, 4) * 100, top_similar.values)
    name_value_pairs = list(zip(top_similar_names, top_similar_values))
    return name_value_pairs
```

```python
author = 'Eminem'
```

```python
find_nearest_authors_cosine(author)
```

```python
find_nearest_authors_euclidean(author)
```
